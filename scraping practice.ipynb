```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/world-population/population-by-country/','html.parser')
response = requests.get(url)
response
```

```python
data = response.text
data
```

```python
soup = BeautifulSoup(data,'lxml')
soup
```

```python
tags = soup.table
tags
```

```python
tag = tags.find_all('tr')
tag
```

```python
tag = tags.find_all('td')
tag
```

```python
a = []
for i in tag:
  s = i.get_text()
  a.append(s)

a
```

```python
final = a[1:]
final
```

```python
for i in final:
    f = i.split('\n')[:]
    print(f)
```

```python
final_data = []
for i in final:
  f = i.split('\n')[:5]
  final_data.append(f)

final_data
```

```python
header = [['Sn','Country','Polutation Increase','Population decrease']]
```

```python
output_data = header+final_data
output_data
```

```python
import csv
f = open('population.csv','w')
write = csv.writer(f)
for i in output_data:
  write.writerow(i)

f.close()
```
